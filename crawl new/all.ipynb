```python
import json

with open("./all.json") as file:
    z = json.load(file)
z
```

```python
new_rs = []
for hotel in z:
    date = hotel['date']
    day,month,year = date.split()
    if int(year)==2023:
        if month in ['Nov',"Dec"]:
            new_rs.append(hotel)
        if month == "Oct":
            if int(day) >= 15:
                new_rs.append(hotel)
```

```python
with open("filter_new.json",'w') as file:
    z = json.dump(new_rs,file, ensure_ascii=False, indent = 2)
```
